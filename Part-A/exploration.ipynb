```python
import matplotlib.pyplot as plt
import seaborn

data = pd.read_csv("/Users/bhavikasewpal/Desktop/CSI4142-Phase-4/Part-A/data/grouped.csv")
data['date'] = pd.to_datetime(data['date'])
fig, ax = plt.subplots(figsize=(20,5))
seaborn.boxplot(x = data.date, 
                y = data.proportionfullyvaccinated, 
                ax = ax)
plt.savefig("graphs/boxplot-proportion-fully-vaccinated.png")
```

```python
fig, ax = plt.subplots(figsize=(20,5))
seaborn.boxplot(x = data.date, 
                y = data.ratechangecases, 
                ax = ax)
plt.savefig("graphs/boxplot-rate-of-change-of-cases.png")
```
